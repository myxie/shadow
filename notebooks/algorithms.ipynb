```python
import os
    # os.chdir('../test/data/shadowgen/')
def setup_dax():
    path = '../test/data/shadowgen/'
    file_list = []
    for file in os.listdir('../test/data/shadowgen/'):
        if 'xml' in file and 'edit' not in file:
            file = path + file
            file_list.append(file)
    file_list
    style = {'description_width': 'initial'}
    w = widgets.Dropdown(
        options=file_list,
    #    value='pineapple', # Defaults to 'pineapple'
    #    layout={'width': 'max-content'}, # If the items' names are long
        description='DAX Workflow File:',
        style=style,
        disabled=False
    )
    
    return w
```

```python
workflow_selection=setup_dax()
display(workflow_selection)
```
